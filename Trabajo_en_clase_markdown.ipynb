#**Informe valor presente neto**

También conocido como Valor Actual Neto (VAN), es un criterio de inversión que consiste en actualizar los cobros y pagos de un proyecto para conocer cuánto se va a ganar o perder con esa inversión

##*Conceptos Clave*
**Valor del dinero en el tiempo** Un peso hoy vale más que un peso mañana debido a la capacidad de generar intereses o a la inflación.

**Tasa de descuento** Es la tasa de retorno mínima exigida para realizar la inversión. En Colombia, por ejemplo, está definida frecuentemente en el 9% para ciertos proyectos.

##*Criterios de Decisión*
Al evaluar un proyecto, los resultados se interpretan así:
- VPN > 0: El proyecto es rentable y se debe aceptar (crea riqueza adicional).
- VPN = 0: No hay ganancia ni pérdida extra; se está al nivel de la tasa de retorno exigida.
- VPN < 0: El proyecto no es rentable y se debe rechazar (destruye valor).

##*formula valor presente neto*
$$VPN = \sum_{t=1}^{n} \frac{R_t}{(1 + i)^t} - I_0$$


- \(I_{0}\): Inversión inicial (salida de dinero en el momento cero).
- \(F_{t}\): Flujo de caja neto en el periodo \(t\).
- \(k\): Tasa de descuento o costo de oportunidad.
- \(n\): Número de periodos (vida útil del proyecto).
